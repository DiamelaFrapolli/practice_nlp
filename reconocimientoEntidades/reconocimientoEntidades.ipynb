```python
nlp = spacy.load("es_core_news_lg")  # Modelo para español
```

```python
doc = nlp("Juan Pérez vive en Buenos Aires.")
for ent in doc.ents:
    if ent.label_ == "PER":
        print(ent.text)
```

```python
# Leer el archivo .txt
with open("nombres.txt", "r", encoding="utf-8") as file:
    text = file.read()
```

```python
# Procesar el texto con spaCy
doc = nlp(text)

# Imprimir los nombres detectados
print("Nombres detectados:")
for ent in doc.ents:
    if ent.label_ == "PER":  # 'PER' indica personas
        print(ent.text)
```

```python
# Leer el archivo .txt
with open("index-2.txt", "r", encoding="utf-8") as file:
    textoCorpus = file.read()
```

```python
# Procesar el texto con spaCy
doc = nlp(textoCorpus)

# Imprimir los nombres detectados
print("Nombres detectados:")
for ent in doc.ents:
    if ent.label_ == "PER":  # 'PER' indica personas
        print(ent.text)
```

```python
for ent in doc.ents:
    if ent.label_ == "PER" and all(token.pos_ in ["PROPN", "NOUN"] for token in ent):
        print(ent.text)
```
